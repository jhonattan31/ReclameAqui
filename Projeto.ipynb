```python
import pandas as pd 
```

```python
df = pd.read_csv('dataset/RECLAMEAQUI_NAGEM.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['TEMPO'] = pd.to_datetime(df['TEMPO'])
```

```python
df['STATUS'].value_counts()
```

```python
def count_palavras(texto):
    return len(texto.split())

df['DESCRICAO'].apply(count_palavras).plot(kind='hist', density=True)
df['DESCRICAO'].apply(count_palavras).plot.kde()
```

```python
# !pip install nltk

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

from wordcloud import WordCloud

# Obter a lista de stopwords em portugu√™s
stopwords_portugues = stopwords.words('portuguese')
```

```python
novas_stopwords = ["produto", "empresa", "comprei", "loja", "n√£o", "pra", "tive", "minha", 
    "dia", "dias", "entrega", "reclame", "aqui", "problema", "j√°", "pois", "outro", "outra", "nagem"]

for palavra in novas_stopwords:
    stopwords_portugues.append(palavra)
```

```python
texto = " ".join(df['DESCRICAO'].astype(str).tolist())
print(len(texto)) 
print(texto[:100])  
```

```python
# Gerar a nuvem de palavras
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=stopwords_portugues,
    colormap='viridis', 
    max_words=50
).generate(texto)
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud - Termos mais frequentes nas reclama√ß√µes')
plt.show()
```

```python
import plotly.express as px

# Gr√°fico 1: S√©rie Temporal do n√∫mero de reclama√ß√µes
# Vamos agrupar por m√™s para a visualiza√ß√£o n√£o ficar muito polu√≠da
reclamacoes_por_mes = df.set_index('TEMPO').resample('M').size().reset_index(name='CONTAGEM')
fig_serie_temporal = px.line(
    reclamacoes_por_mes, 
    x='TEMPO', 
    y='CONTAGEM', 
    title='S√©rie Temporal de Reclama√ß√µes (por m√™s)',
    labels={'TEMPO': 'Data', 'CONTAGEM': 'N√∫mero de Reclama√ß√µes'}
)

# Gr√°fico 2: Frequ√™ncia de reclama√ß√µes por estado
reclamacoes_por_estado = df['LOCAL'].str.split(' - ').str[-1].value_counts().reset_index()
fig_por_estado = px.bar(
    reclamacoes_por_estado, 
    x='LOCAL', 
    y='count', 
    title='Reclama√ß√µes por Estado',
    labels={'LOCAL': 'Estado', 'count': 'N√∫mero de Reclama√ß√µes'}
)

# Gr√°fico 3: Frequ√™ncia por tipo de STATUS
reclamacoes_por_status = df['STATUS'].value_counts().reset_index()
fig_por_status = px.pie(
    reclamacoes_por_status, 
    names='STATUS', 
    values='count', 
    title='Distribui√ß√£o por Status da Reclama√ß√£o',
    hole=.3 # Para criar um gr√°fico de rosca (donut chart)
)

# Gr√°fico 4: Distribui√ß√£o do tamanho dos textos (vers√£o Plotly)
df['TAMANHO_TEXTO'] = df['DESCRICAO'].apply(lambda x: len(x.split()))
fig_dist_texto = px.histogram(
    df, 
    x='TAMANHO_TEXTO', 
    nbins=50, 
    title='Distribui√ß√£o do Tamanho das Reclama√ß√µes (em palavras)',
    labels={'TAMANHO_TEXTO': 'Quantidade de Palavras'}
)
```
